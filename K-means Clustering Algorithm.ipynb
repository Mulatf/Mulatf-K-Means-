# K-Means clustring
Here i will applied k-means clustring algorithms on iris dataset and custome dataset.

```python
# Import wanted libs
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
# Load iris dataset
iris = load_iris()
X = iris.data
```

```python
# Visualize the data before clustering
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c='blue')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Original Data')
```

```python
# Apply k-means clustering
k = 3
kmeans = KMeans(n_clusters=k)
kmeans.fit(X)
```

```python
# Get cluster assignments and cluster centers
labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

```python
# Visualize the data after clustering
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('K-means Clustering')

plt.tight_layout()
plt.show()
```

So as you see digram above the clusters is overlapping, so we go to fix that using PCA (Principal Component Analysis), and do this work again using custom data.

```python
# Import PCA
from sklearn.decomposition import PCA
```

```python
# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
# Visualize the data after clustering
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels)
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-means Clustering (PCA-reduced Data)')
plt.show()
```

As you see above digram, the above digram get enchenced.

```python
# Import make_blobs
from sklearn.datasets import make_blobs
```

```python
# Generate sample data
n_samples = 200
random_state = 42
X1, y = make_blobs(n_samples=n_samples, random_state=random_state) # make sure here we assign our generated data to "X1" not "X"
```

```python
# Visualize the data before clustering
fig = plt.figure(0)
plt.grid(True)
plt.scatter(X1[:,0],X1[:,1], c = "blue")
plt.title('Before Clustering')
plt.show()
```

```python
# Apply K-means clustering
k = 3
kmeans = KMeans(n_clusters=k, random_state=random_state)
kmeans.fit(X1) # we use "X1"
```

```python
# Get cluster assignments and cluster centers
labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

```python
# Visualize the data and cluster centers
plt.scatter(X1[:, 0], X1[:, 1], c=labels) #We use "X1"
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('K-means Clustering')
plt.show()
```
